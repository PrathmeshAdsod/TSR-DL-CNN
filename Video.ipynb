```python
import cv2
import numpy as np
import pickle
```

```python
frameWidth = 640
frameHeight = 480
Brightness = 180
threshold = 0.90
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
```

```python
cap = cv2.VideoCapture(0)
cap.set(3, frameWidth)
cap.set(4, frameHeight)
cap.set(10, Brightness)
```

```python
from tensorflow import keras
my_model = keras.models.load_model('my_model.h5')
my_model.load_weights('model_weights.h5')
```

```python
def grayscale(img) :
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img

def equalize(img) :
    img = cv2.equalizeHist(img)
    return img


def preprocessing(img) :
    img = grayscale(img)
    img = equalize(img)
    img = img / 255
    # print(img)
    return img
```

```python
classes = { 1:'Speed limit (20km/h)',
            2:'Speed limit (30km/h)', 
            3:'Speed limit (50km/h)', 
            4:'Speed limit (60km/h)', 
            5:'Speed limit (70km/h)', 
            6:'Speed limit (80km/h)', 
            7:'End of speed limit (80km/h)', 
            8:'Speed limit (100km/h)', 
            9:'Speed limit (120km/h)', 
            10:'No passing', 
            11:'No passing veh over 3.5 tons', 
            12:'Right-of-way at intersection', 
            13:'Priority road', 
            14:'Yield', 
            15:'Stop', 
            16:'No vehicles', 
            17:'Veh > 3.5 tons prohibited', 
            18:'No entry', 
            19:'General caution', 
            20:'Dangerous curve left', 
            21:'Dangerous curve right', 
            22:'Double curve', 
            23:'Bumpy road', 
            24:'Slippery road', 
            25:'Road narrows on the right', 
            26:'Road work', 
            27:'Traffic signals', 
            28:'Pedestrians', 
            29:'Children crossing', 
            30:'Bicycles crossing', 
            31:'Beware of ice/snow',
            32:'Wild animals crossing', 
            33:'End speed + passing limits', 
            34:'Turn right ahead', 
            35:'Turn left ahead', 
            36:'Ahead only', 
            37:'Go straight or right', 
            38:'Go straight or left', 
            39:'Keep right', 
            40:'Keep left', 
            41:'Roundabout mandatory', 
            42:'End of no passing' }
```

```python
while True :
    success, imgOriginal = cap.read()
    img = np.asarray(imgOriginal)
    
    img = cv2.resize(img , (30,30)) #img = cv2.resize(img , (32,32))
    # We have not converted data into GraScale form when we trained our mode
    # So we should not preprocess it here otherwise our dimensions change
    #img = preprocessing(img)
    cv2.imshow("Image ", img)
    img = np.reshape(img, (1, 30, 30, 3))  # Reshape to (1, 30, 30, 3)
  
    
    #cv2.putText(imgOriginal, 'TRAFFIC SIGN NAME :  ', (100,50), font, 0.75, (0,255,0), 2, cv2.LINE_AA)
    #cv2.putText(imgOriginal, 'PROBABILITY : ', (110,90), font, 0.75, (0,255,0), 2, cv2.LINE_AA)
    
    predictions = my_model.predict(img)
   # classIndex = my_model.predict_classes(img)
    #classIndex = np.argmax(predictions)
    
    classIndex = np.argmax(predictions)
    print(classIndex+1)
    probabilityValue = np.amax(predictions)
    print(probabilityValue * 100)
    
    if probabilityValue > threshold :
        cv2.putText(imgOriginal, str(classIndex+1) + '  ' + classes[1+classIndex] , (150,50), font, 0.75, (0,255,0), 2, cv2.LINE_AA)
        cv2.putText(imgOriginal, str(probabilityValue * 100) , (140,90), font, 0.75, (0,255,0), 2, cv2.LINE_AA)

    cv2.imshow('Result ', imgOriginal)
    
    if cv2.waitKey(1) & 0xFF == ord('q') :
        break
        
cap.release()
cv2.destroyAllWindows()
```
